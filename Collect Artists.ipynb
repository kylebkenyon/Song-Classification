```python
import string
import numpy
import requests
import urllib.request
import json
from bs4 import BeautifulSoup
import re
```

```python
BASE_URL = "https://www.besteveralbums.com/bandstats.php?l=0&showchildren=0&y=0&d=0&page="
```

```python
def getArtistsFromPage(page_number):
    url = BASE_URL + str(page_number)
#     print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    
    artists = []
    nav2_tags = soup.find_all(class_='nav2')
    for i in range(len(nav2_tags) - 1):
        if (i % 2 == 0):
            artists.append(nav2_tags[i].text)
    return artists
```

```python
final_artists = []
for page in range(1,51):
    page_artists = getArtistsFromPage(page)
    for artist in page_artists:
        final_artists.append(artist)
```
